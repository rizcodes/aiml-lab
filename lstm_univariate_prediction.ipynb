# Lets consider predicting sequence using LSTM
[10, 20, 30, 40, 50, 60, 70, 80, 90]

### Data Prep -  Initializing X and Y

```python
import numpy as np
```

```python
# Initialize an array 'x' 6 rows * 3 columns
x = np.empty([6,3])
print (x)
```

```python
# Add the data sequence in array x
a = 10
m = 1
for i in range(len(x)):
    n = a * m
    for j in range(3):
        x[i][j] = n
        n += 10
    m += 1
    
print (x)
```

```python
# Initialize array for y (labels) 6 rows * 1 cols
y = np.empty([6,1])
print (y)
```

```python
# Assign labels in y
n = 40
for i in range(len(y)):
    y[i] = n
    n += 10

print (y)
```

```python
x.shape, y.shape
```

### Define LSTM

```python
#Import modules
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
```

```python
#Define input feature
n_steps = 3              #number of time steps - [10 20 30] = [40]
n_features = 1            #Univariate - hence only one feature
```

```python
#Define model
model = Sequential()
model.add(LSTM(units=50, activation='relu', input_shape=(n_steps, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

## Train the model

```python
#We have multiple samples, hence model will expect input data to have dimension as
#[n_samples, n_steps, n_features]
#In our case we have 
# n_samples: 6
# n_steps: 3
# n_features: 1
```

```python
#Input shape
x.shape
```

```python
#Need to re-shape as (6, 3, 1)
X = x.reshape((x.shape[0], x.shape[1], n_features))
X.shape
```

```python
# Fit model
model.fit(X, y, epochs=256, verbose=1)
```

## Test Model
Test sequence: [70, 80 ,90] = [100]

```python
#Initialize test array
x_test = np.array([70, 80, 90])
x_test
```

```python
#Reshape test input
X_test = x_test.reshape(1, n_steps, n_features)
X_test.shape
```

```python
#Predict output
yhat = model.predict(X_test, verbose=1)
```
